## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
filtered_data = df[['age', 'pclass', 'sex']]
```

```python

sns.boxplot(x='sex', y='age', hue='pclass', data=filtered_data)
plt.title('Distribution of Males and Females by Age and Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Show the plot
plt.show()
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
columns_of_interest = ['sibsp', 'parch', 'age', 'fare', 'survived']
correlation_matrix = df[columns_of_interest].corr()
plt.figure(figsize=(10, 8))  # Adjust the figure size as per your preference
sns.heatmap(correlation_matrix, annot=True,  fmt=".2f", linewidths=0.5)
plt.title('Titanic Dataset')
plt.show()
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
# Draw the factorplot
g= sns.catplot(x='sibsp', y='survived', data=df, size=5, aspect=1.5)

# Show the plot
plt.show()
```
